```python
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import random as rand
%matplotlib inline
```

```python
rand.randint(1,6)
```

```python
data = ([20,40,60],[2,10,2])
# all_data = initial_data
point = [35, 6]
for n in range(1,100):
    x, y = point[0], point[1]
    r = rand.randint(1,6)
    i = 0
    res_x = int((point[0]-data[0][i])/2)
    res_y = int((point[1]-data[1][i])/2)
    if r == 1 or r == 2:
        i = 1
    if r == 3 or r == 4:
        i = 2
    if r == 5 or r == 6:
        i = 0
    if i == 0:
        t_x = x-res_x 
        t_y = y-res_y
        if t_x > data[0][i]:
            x -= t_x
        if t_y > data[1][i]:
            y -= res_y 
    if i == 1:
        t_x = x+res_x 
        t_y = y+res_y
        if t_x < data[0][i]:
            x += t_x
        if t_y < data[1][i]:
            y += res_y  
    if i == 2:
        t_x = x+res_x 
        t_y = y-res_y
        if t_x < data[0][i]:
            x += t_x
        if t_y > data[1][i]:
            y -= res_y  
    print(i,x,y, point, data)
    point[0] = x
    point[1] = y
    data[0].append(point[0])
    data[1].append(point[1])
    
```
