```python
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import random as rand
%matplotlib inline
```

```python
rand.randint(1,6)
```

```python
data = ([rand.randint(1,100),rand.randint(1,100),rand.randint(1,100),rand.randint(1,100),rand.randint(1,100)],[rand.randint(1,100),rand.randint(1,100),rand.randint(1,100),rand.randint(1,100),rand.randint(1,100)])
# all_data = initial_data
point = [0, 0]
for n in range(1,10000):
    x, y = point[0], point[1]
    r = rand.randint(1,5)
    i = 0
    if r == 1:
        i = 0
    if r == 2:
        i = 1
    if r == 3:
        i = 2
    if r == 4:
        i = 3
    if r == 5:
        i = 4
    res_x = abs((point[0]-data[0][i])/2)
    res_y = abs((point[1]-data[1][i])/2)
    x += res_x if point[0] < data[0][i] else -res_x
    y += res_y if point[1] < data[1][i] else -res_y
    point[0] = x
    point[1] = y
    data[0].append(point[0])
    data[1].append(point[1])
    
```

```python
def draw_point():
    
    plt.scatter(data[0], data[1], s=1)
    plt.show()

if __name__ == '__main__':
    draw_point()
```
